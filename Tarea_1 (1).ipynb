Investigue y explique brevemente la relación **teórica** entre el `GDP (% per capita)` y cada una de las variables denominadas como "Características del país" en la introducción.

Ejemplo: Existe una variable denominada `Coastline (coast/area ratio)`. Coastline es una medida de la cantidad de costa (acceso a mar) del país normalizada al área total del país para no beneficiar a países más grandes pero con la misma proporción de costa. A mayor "Costline" aumenta la capacidad portuaria per capita del país, más puertos facilita el comercio y podría aumentar el GDP per cápita.




---

*   Population:

 La "Population", que hace referencia a la cantidad de personas en un país, tiene relación directa e inversamente proporcional al GDP **per cápita**. El cálculo del GDP per cápita es igual a: GDP/Population. Así, a mayor "Population" y ante un GDP constante, baja el GDP per cápita. Por otra parte, si la "Population" baja a un GDP constante, el GDP per cápita sube.

 Si consideramos un crecimiento del PIB a una tasa más alta que el crecimiento de la población ("Population"), el GDP per cápita crece.

*   Area (sq. mi.):
  
   El área de un país en millas cuadradas ("Área (sq. mi.)") no afecta directamente al PIB per cápita ya que no se utiliza la medida de forma directa para su cálculo. Sin embargo, esta variable puede influir en la eficiencia de su economía, en la disponibilidad y explotación de recursos, en los costos de infraestructura y desarrollo y en la distribución de la población. Estas variables afectan directamente a la generación de ingresos en un país en distintas magnitudes, por lo que podemos distinguir una relación entre ambos.

   A mayor área, con una "Population" constante, el efecto en el PIB dependerá de la cápacidad de explotación y la calidad de los recursos en dicha área.

*   Pop. Density (per sq. mi.)
  
   La densidad de la población por milla cuadrada, al igual que la variable "Área", no tiene una relación directa, sin embargo tiene efecto en variables económicas que afectan a la generación del PIB, como lo es la eficiencia económica, la urbanización, infraestuctura, tencnología.

   Así como la variable anterior, el efecto en el PIB dependerá de otras variables de la economía.

*   Coastline (coast/area ratio):
    
   Coastline es una medida de la cantidad de costa (acceso a mar) del país normalizada al área total del país para no beneficiar a países más grandes pero con la misma proporción de costa. A mayor "Costline" aumenta la capacidad portuaria per capita del país, más puertos facilita el comercio y podría aumentar el GDP per cápita.

*   Net migration:
    
   Esta variable corresponde a Imigrantes - Emigrantes, si el componente es positivo, el neto de personas en un país aumenta. Por otra parte, si este es negativo, el neto de personas en un país, disminuye.

   Esta variable puede tener efectos en el PIB con un efecto similar a la variable "Population" y podemos distinguir una correlación positiva entre ambas variables. Cuando "Net migraton" es positiva, si el PIB no crece a la misma tasa que el crecimiento de la "Net migration", el PIB per cápita baja. La maginitud de la relación directa e inversamente proporcional por ende, dependerá de variables como el talento e impacto que genera este aumento de migración con la inversión extranjera (A e I, variables presentes en el cálculo del PIB).

*   Infant mortality (per 1000 births):

  Esta variable, afecta directamente al PIB per cápita, y al igual que las variables anteriores, correlaciona positivamente con la variable "Population". A mayor "Infant mortality", la "Population" baja siempre que la natalidad se mantenga constante, y ante esto, el GDP per cápita baja.  

*   GDP ($ per capita):
    
   El cálculo del GDP per cápita corresponde al GDP de un país dividido por la cantidad de personas qu hay en el país, es decir GDP/"Population". El GDP o PIB se puede calcular de distintas formas, una de ellas es el metodo del gasto, y se obtiene a partir de la suma de C+I+G+(X−M); C es igual a consumo, I corresponde al nivel de inversión, G corresponde al gasto de gobierno y (X-M) corresponde a las exportaciones netas.

*   Literacy (%): La variable se relaciona al GDP ($ per capita) mediante la variable de aprendizaje, ya que la optimización de la producción de un país se correlaciona fuertemente con esto.

*   Phones (per 1000): Esta variable se relaciona al nivel de digitalización de un país, lo que se correlaciona directamente con la tecnología y la forma en la que se optimiza la producción, por lo que podemos esperar una correlación positiva entre esta variable y el PIB per cápita.

*   Arable (%): Tiene un comportamiento similar a la variable "Area", ya que corresponde al espacio explotable para producción, es decir el producto potencial de un país en relación a la agriculura. Sin embargo, podríamos inferir que en paises con menor PIB per cápita, la industrialización sea más baja y el enfoque en la agricultura sea mayor. Por otra parte, en paises con PIB per capita más alto, el enfoque sea mayor en producciones industriales y menor en agricultura. Sin embargo, esta variable puede estar más correlacionada con el área geográfica y con el nivel de industrialización de un país.

*   Crops (%):  Representa el porcentaje de la tierra dedicada a cultivos agrícolas, puede tener una relación inversa o positiva con el PIB per cápita de un país. En países con un PIB per cápita bajo, es más probable que una mayor proporción del uso de la tierra se destine a la agricultura debido a una menor industrialización y desarrollo económico. En cambio, en países con un PIB per cápita alto, una menor proporción de tierra se dedica a cultivos, ya que los sectores industrial y de servicios tienden a ser más predominantes, reflejando una economía más diversificada y menos dependiente de la agricultura. Así, la relación entre Crops (%) y PIB per cápita puede variar dependiendo del nivel de desarrollo económico del país.

*   Other (%): otros factores pueden tener correlaciones diversas según sea lo que esto considere.

*   Climate: Puede tener una relación indirecta con el PIB per cápita a través de su impacto en la agricultura y el desarrollo económico. En países con un clima favorable para la agricultura, es posible que haya un mayor uso de tierras arables, lo que podría condicionar a un país a que sus actividades sean más relacionadas a la agricultura. Sin embargo, en países con climas menos favorables, la dependencia de la agricultura puede ser limitada, lo que impulsa el desarrollo de sectores industriales y de servicios. Esto puede impulsar a PIBs per cápita más altos.

*   Birthrate: Un mayor PIB per cápita está típicamente asociado con una menor Birthrate, ya que la mayor prosperidad y el acceso a educación y salud suelen reducir las tasas de natalidad.

*   Deathrate: Un mayor PIB per cápita está típicamente asociado con una Deathrate más baja, ya que el desarrollo económico contribuye a mejores condiciones de salud y mayor longevidad

*   Agriculture: La relación suele ser inversa. En países con un PIB per cápita bajo, una mayor proporción del ingreso nacional proviene de la agricultura, ya que estas economías dependen más de actividades agrícolas. En cambio, en países con PIB per cápita alto, la agricultura representa una menor proporción del PIB, ya que estos países tienen economías más diversificadas, con un enfoque mayor en la industria y los servicios.

*   Industry: La relación tiende a ser positiva. En países con un PIB per cápita bajo, una menor proporción del PIB proviene de la industria, ya que las economías están más centradas en la agricultura. En países con un PIB per cápita alto, la Industry representa una mayor proporción del PIB, ya que estos países han experimentado un proceso de industrialización que impulsa su crecimiento económico, con sectores como la manufactura, la construcción y la minería jugando un papel clave.

*   Service: La relación suele ser positiva. En países con un PIB per cápita bajo, el sector de los servicios representa una menor proporción del PIB, ya que estas economías están más centradas en la agricultura o la industria. A medida que el PIB per cápita crece, los países suelen experimentar un cambio estructural hacia una mayor participación de los servicios en la economía, impulsado por sectores como tecnología, salud, educación, finanzas y turismo.

### Pregunta 1.10

Supongamos que `GDP ($ per capita)` es su variable objetivo. Estudie la correlación de esta variable con el resto de las variables del `df`. ¿Por qué es importante analizar la correlación entre las variables?



```python
# Calcular la matriz de correlación (solo con columnas numéricas)
correlation_matrix = df[df.dtypes[df.dtypes != "object"].index].corr()

# Filtrar solo la correlación con GDP ($ per capita)
gdp_correlation = correlation_matrix[['GDP ($ per capita)']].sort_values(by='GDP ($ per capita)', ascending=False)

# Mostrar las correlaciones
print(gdp_correlation)

# Graficar la correlación con un heatmap
plt.figure(figsize=(6, 8))
sns.heatmap(gdp_correlation, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlación del PIB per cápita con otras variables")
plt.show()

```


---


Analizar la correlación de las variables con la variable que se busca estimar (variable objetivo) es de suma relevancia, ya que dependiendo del nivel de correlaciones entre variables y con la variable objetivo es que ellas serán buenos predictores del GDP per cápita. En este caso, Phones (per 1000), Service y Literacy son las variables que más correlacionan con el GDP per cápita.  En este caso, las variables que buscan explicar a la variable objetivo pueden estar correlacionando entre si generando problemas de colinealiedad. Dicha variación porcentual puede ser positiva o negativa, pero ordene los el dataframe de tal manera que la variación de correlación absoluta sea desendiente.

```python
# Filtrar solo columnas numéricas
df_numeric = df.select_dtypes(include=['number'])

# Calcular la correlación de GDP ($ per capita) con todas las columnas
corr_per_capita = df_numeric.corr()['GDP ($ per capita)'].abs()

# Calcular la correlación de GDP ($) con todas las columnas
corr_total_gdp = df_numeric.corr()['GDP'].abs()

# Crear un nuevo DataFrame con la variación porcentual
correlation_variation = pd.DataFrame({
    'Correlation GDP per capita': corr_per_capita,
    'Correlation GDP': corr_total_gdp,
    'Percentage Variation': ((corr_total_gdp - corr_per_capita) / corr_per_capita) * 100
})

# Ordenar en orden descendente
correlation_variation_sorted = correlation_variation.sort_values(by='Percentage Variation', ascending=False)

# Aplicar formato a la tabla
styled_table = correlation_variation_sorted.style.format({
    'Correlation GDP per capita': "{:.3f}",
    'Correlation GDP': "{:.3f}",
    'Percentage Variation': "{:+.2f}%"
}).set_caption("Correlation Variation Between GDP ($ per capita) and GDP ($)")

# Mostrar la tabla
display(styled_table)

```

### Pregunta 2.4

Para el año 2007, por cada país realice un ranking de las fuentes con más emisiones de CO2 excluyendo las variables `Per Capita` y `Bunker fuels (Not in Total)`. Es decir, asigne un número de 1 a 5 a $\{$ `Solid Fuel`, `Liquid Fuel`, `Gas Fuel`, `Cement`, `Gas Flaring` $\}$, donde 1 es la mayor fuente de emisión de ese país en aquel año, y 5 indica que fue la menor; así para todos los países.

Si en 2007 no se reporta una fuente de emisión para un país, por ejemplo, si emisiones de `Gas Flaring` no se reportara, entonces asigne números de 1 a 4 a las fuentes restantes. Análogo para un menor número de datos.

Luego, por cada variable grafique un histograma de frecuencias del ranking que obtuvo la fuente emisión a lo largo de todos los países.

¿Cuál fue la fuente más contaminante en la mayoría de países en 2007?


```python
# Definir las fuentes de emisión relevantes
emission_sources = ['Solid Fuel', 'Liquid Fuel', 'Gas Fuel', 'Cement', 'Gas Flaring']

# Filtrar los datos para el año 2007
dfco2_2007 = df_co2[df_co2['Year'] == 2007]

# Crear un nuevo DataFrame para almacenar los rankings
rankings = pd.DataFrame(columns=emission_sources, index=dfco2_2007['Country'])

# Asignar el ranking de emisiones para cada país
for country in dfco2_2007['Country']:
    # Filtrar las emisiones del país
    country_data = dfco2_2007[dfco2_2007['Country'] == country]

    # Filtrar las fuentes de emisión relevantes (excluyendo 'Per Capita' y 'Bunker fuels (Not in Total)')
    emissions = country_data[emission_sources].iloc[0]

    # Eliminar las fuentes con valor NaN
    emissions = emissions.dropna()

    # Asignar ranking
    ranking = emissions.sort_values(ascending=False).rank(method='min', ascending=False)

    # Asignar el ranking al DataFrame
    rankings.loc[country, emissions.index] = ranking

# Mostrar el DataFrame con los rankings
print(rankings)

# Graficar histogramas de frecuencias de los rankings
plt.figure(figsize=(12, 8))

for idx, source in enumerate(emission_sources):
    plt.subplot(2, 3, idx + 1)  # Usar subgráficos para cada fuente
    plt.hist(rankings[source], bins=range(1, 7), edgecolor='black', align='left')
    plt.title(f'Ranking Frequencies: {source}')
    plt.xlabel('Ranking')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Calcular cuál fue la fuente más contaminante en la mayoría de países en 2007
most_contaminating_source = rankings.mode(axis=0).iloc[0]
print("Fuente más contaminante en la mayoría de los países en 2007:")
print(most_contaminating_source)
```

A partir de los resultados, vemos que la fuente mas alta de contaminantes es el Liquid Fuel con una puntuación de 1 en promedio.
